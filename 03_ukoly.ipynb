Vypište první tři prvky seznamu `[1, 2, 3, 4, 5]`.


````python
seznam = [1, 2, 3, 4, 5]
for i, cislo in enumerate(seznam):
    if i == 3:
        break
    print(cislo)
````

## 1.6 For cyklus s `else`
Vypište první tři prvky seznamu. Pokud je seznam dlouhý nejvýše 3 prvky, vypište na závěr `"Krátký seznam!"`.

Vyzkoušejte na `[1, 2, 3, 4, 5]` a `[1, 2]`.


````python
def vypis(seznam):
    for i, cislo in enumerate(seznam):
        if i == 3:
            break
        print(cislo)
    else:
        print("Krátký seznam!")
vypis([1, 2, 3, 4, 5])
vypis([1, 2])
````

````python
seznam = [1, 2, 3, 4, 5]
for cislo in seznam:
    if cislo % 2 != 0:
        continue
    print(cislo)
````

## 1.11 While cyklus
Při počátečním vkladu 10 000 Kč, roční úrokové sazbě 9 % (úrok se připíše na konci roku) a ročním vkladu 1 000 Kč na začátku každého roku určete, za kolik let bude na účtu 1 000 000 Kč.


````python
balance = 10000
vklad = 1000
urok = 1.09 # 100% + 9% = 109% nebo 109/100 = 1.09
target_balance = 1000000
years = 0

while balance < target_balance:
    years += 1

    balance += vklad
    balance *= urok

print(years)
print(balance)
````

# 2. Generátory a iterátory
## 2.1 Fibonacciho generátor
Napište generátor (funkci), který vytváří Fibonacciho posloupnost a postupně vrátí prvních `n` členů. Otestujte pomocí `for` cyklu.


````python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

pocet = 100
for cislo in fibonacci(pocet):
    print(cislo, end=" ")
````

## 2.2 Generátorový výraz
Napište generátorový výraz, který vrací všechna sudá čísla menší než 100. Otestujte pomocí `for` cyklu.


````python
suda_cisla = (x for x in range(100) if x % 2 == 0)
for cislo in suda_cisla:
    print(cislo, end=" ")
````

## 2.3 Test dělitelnosti pomocí generátoru
Pro dané číslo `n` (berte `n >= 2`) vytvořte generátor, který pro každé `d` od `2` do `n-1` vrací výraz `n % d != 0`. Otestujte pro několik čísel a použijte `all()` tak, aby výraz fungoval jako test prvočíselnosti.


````python
def prime(n):
    if n < 2: return False
    return all(n % d != 0 for d in range(2, n))

print(prime(7))
print(prime(10))
````

## 4.1 Vytvoření modulu přes `%%writefile`
Použijte `%%writefile` a vytvořte soubor `factorial.py` se dvěma funkcemi:

```python
def factorial_loop(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
```

Modul načtěte a zavolejte obě funkce.


````python
%%writefile factorial.py
def factorial_loop(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
````

````python
import factorial
print(factorial.factorial_loop(5))
print(factorial.factorial_recursive(5))
````

## 5.2 Palindrom
Vytvořte funkci `je_palindrom`, která vrátí `True`, pokud je zadaný řetězec palindrom, jinak `False`. Otestujte ji na několika příkladech.


````python
def je_palindrom(word):
    new_word = word.replace(" ", "").lower()
    palindrom = ""
    for i in range(len(new_word) - 1, -1, -1):
        palindrom += new_word[i]
    return new_word == palindrom
print(je_palindrom("Level"))
print(je_palindrom("Python"))
print(je_palindrom("step on no pets"))
````
