13. Vytvořte strukturované pole s koordináty `x` a `y` pokrývající oblast `[0, 1] x [0, 1]`.


````python
grid_size = 5
struct_array = np.zeros((grid_size, grid_size), [('x', float), ('y', float)])
x_coords, y_coords = np.meshgrid(np.linspace(0, 1, grid_size), np.linspace(0, 1, grid_size))
struct_array['x'] = x_coords
struct_array['y'] = y_coords
print(struct_array[:2, :2])
````

2. Vytvořte tři náhodné matice 5x5 a uložte je do souboru ve formátu HDF5.


````python
import h5py
````

````python
mat1 = np.random.rand(5, 5)
mat2 = np.random.rand(5, 5)
mat3 = np.random.rand(5, 5)\

with h5py.File('matice.h5', 'w') as hdf_file:
    hdf_file.create_dataset('matrix_1', data=mat1)
    hdf_file.create_dataset('matrix_2', data=mat2)
    hdf_file.create_dataset('matrix_3', data=mat3)
````

3. Načtěte soubor z předchozího úkolu, zobrazte obsah každé matice a zjistěte jejich rozměry.


````python
with h5py.File('matice.h5', 'r') as hdf_file:
    for key in hdf_file.keys():
        loaded_data = hdf_file[key][:]
        print(key)
        print(loaded_data.shape)
        print(loaded_data)
````
