1. Vynásobte matici 5x3 maticí 3x2 (maticový součin).


````python
A = np.random.rand(5, 3)
B = np.random.rand(3, 2)
C = A @ B 
C
````

2. Vezměte náhodnou matici 10x2 představující kartézské souřadnice a převeďte je na polární souřadnice.

````python
Z = np.random.random((10, 2))
X, Y = Z[:, 0], Z[:, 1]
R = np.sqrt(X**2 + Y**2)
T = np.arctan2(Y, X)
print(R)
print(T)
````

3. Pro dvě pole `X` a `Y` zkonstruujte Cauchyho matici `C` (`C_ij = 1 / (x_i - y_j)`).

````python
X = np.arange(5)
Y = X + 0.5
C = 1.0 / (X[:, np.newaxis] - Y)
C
````

4. Uvažujte náhodné pole tvaru `(100, 2)` reprezentující souřadnice a spočítejte vzdálenost pro každý bod.

````python
Z = np.random.random((100, 2))
D = cdist(Z, Z)
print(D.shape)
````

5. Ověřte empiricky na náhodných maticích, že platí $(AB)^T = B^T A^T$.

````python
A = np.random.rand(3, 3)
B = np.random.rand(3, 3)

leva_strana = (A @ B).T
prava_strana = B.T @ A.T

np.allclose(leva_strana, prava_strana)
````

6. Pro náhodné 1D vektory $u, v$ vypočítejte dyadický součin $uv$, kde $$(uv)_{ij} = u_i v_j$$, pomocí `newaxis`. Zkontrolujte výsledek s `np.outer(u, v)`.

````python
u = np.random.rand(5)
v = np.random.rand(5)

vysledek = u[:, np.newaxis] * v
kontrola = np.outer(u, v)

np.allclose(vysledek, kontrola)
````

7. Vypočítejte hodnost matice o velikosti 5x6 naplněné hodnotami 0, 1, 2, 3 ... Byli byste schopni zdůvodnit, proč je hodnost právě taková?

4. Jaký je výsledek následujícího výrazu?

````python
print(0 * np.nan) # nan
print(np.nan == np.nan) # False
print(np.inf > np.nan) # False
print(np.nan - np.nan) # nan
print(np.nan in set([np.nan])) # True
print(0.3 == 3 * 0.1)  # False
````
