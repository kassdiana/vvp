10. Upravte funkci nalézající prvočísla tak, aby ukládala už spočítaná prvočísla do listu v defaultním argumentu funkce.

- Funkce má uchovávat už spočítaná prvočísla v defaultním argumentu.
- Při dalším volání má dopočítat jen chybějící prvočísla.
- Vracejte vždy **kopii** požadované části seznamu.


````python
def najdi_n_prvocisel(n):
    prvocisla = []
    i = 2
    while len(prvocisla) < n:
        je_prvocislo = True
        for prvocislo in prvocisla:
            if i % prvocislo == 0:
                je_prvocislo = False
                break
        if je_prvocislo:
            prvocisla.append(i)
        i += 1
    return prvocisla

najdi_n_prvocisel(10)
````

````python
def najdi_n_prvocisel(n, ulozeno=[]):
    if len(ulozeno) == 0:
        i = 2
    else:
        i = ulozeno[-1] + 1
        
    while len(ulozeno) < n:
        je_prvocislo = True
        
        for p in ulozeno:
            if i % p == 0:
                je_prvocislo = False
                break
                
        if je_prvocislo:
            ulozeno.append(i)
            
        i += 1
    return ulozeno[:n]
````

2. Vytvořte třídu implementující metodu `__call__`, která bude mít stejnou funkčnost jako funkce pro vypisování prvních `n` prvočísel (s ukládáním do defaultního argumentu).


````python
class Prvocisla:
    def __init__(self):
        self.argument = []

    def _pocitat_pocet(self, n):
        if len(self.argument) == 0:
            i = 2
        else:
            i = self.argument[-1] + 1
            
        while len(self.argument) < n:
            for p in self.argument:
                if i % p == 0:
                    break
            else:
                self.argument.append(i)
            i += 1

    def __call__(self, n):
        self._pocitat_pocet(n)
        return self.argument[:n]
    
generator = Prvocisla()
print(generator(5))
````

3. Rozšiřte třídu o metodu `__getitem__`, která bude vracet prvočísla podle indexu (podporujte celočíselné indexy i indexy ve formě `slice`).

- Hodit se vám bude `isinstance` pro rozlišení, zda je vstup `slice` nebo `int`.
- Připomeňte si, že objekt typu `slice` má atributy `start`, `stop` a `step`.


````python
def __getitem__(self, index):
    if isinstance(index, int):
        pocet = index + 1 if index >= 0 else abs(index)
        self._pocitat_pocet(pocet)
        return self.argument[index]
        
    elif isinstance(index, slice):
        if index.stop is not None:
            self._pocitat_pocet(index.stop)
        return self.argument[index]
    else:
        raise TypeError()
            
Prvocisla.__getitem__ = __getitem__
print(generator[4])
print(generator[0:3])
````

```python
class Zlomek:
    def __init__(self, citatel, jmenovatel):
    def __add__(self, other):
    def __sub__(self, other):
    def __mul__(self, other):
    def __truediv__(self, other):
    def simplify(self):
    def __str__(self):
    def __repr__(self):
    def to_float(self):
    def __le__(self, other):
    def __lt__(self, other):

    def __gt__(self, other):
```

````python
class Zlomek:
    def __init__(self, citatel, jmenovatel):
        if jmenovatel == 0:
            raise ValueError("Jmenovatel nesmí být nula.")
        self.citatel = citatel
        self.jmenovatel = jmenovatel
        self.simplify()

    def simplify(self):
        delitel = math.gcd(self.citatel, self.jmenovatel)
        self.citatel //= delitel
        self.jmenovatel //= delitel
        
        if self.jmenovatel < 0:
            self.citatel = -self.citatel
            self.jmenovatel = -self.jmenovatel

    def __add__(self, other):
        novy_citatel = (self.citatel * other.jmenovatel) + (other.citatel * self.jmenovatel)
        novy_jmenovatel = self.jmenovatel * other.jmenovatel
        return Zlomek(novy_citatel, novy_jmenovatel)

    def __sub__(self, other):
        novy_citatel = (self.citatel * other.jmenovatel) - (other.citatel * self.jmenovatel)
        novy_jmenovatel = self.jmenovatel * other.jmenovatel
        return Zlomek(novy_citatel, novy_jmenovatel)

    def __mul__(self, other):
        return Zlomek(self.citatel * other.citatel, self.jmenovatel * other.jmenovatel)

    def __truediv__(self, other):
        if other.citatel == 0:
            raise ZeroDivisionError("Nelze dělit nulou")
        return Zlomek(self.citatel * other.jmenovatel, self.jmenovatel * other.citatel)

    def __str__(self):
        if self.jmenovatel == 1:
            return str(self.citatel)
        return f"{self.citatel}/{self.jmenovatel}"

    def __repr__(self):
        return f"Zlomek({self.citatel}, {self.jmenovatel})"

    def to_float(self):
        return self.citatel / self.jmenovatel
    def __eq__(self, other):
        return (self.citatel * other.jmenovatel) == (other.citatel * self.jmenovatel)

    def __lt__(self, other):
        return (self.citatel * other.jmenovatel) < (other.citatel * self.jmenovatel)

    def __le__(self, other):
        return (self.citatel * other.jmenovatel) <= (other.citatel * self.jmenovatel)

    def __gt__(self, other):
        return (self.citatel * other.jmenovatel) > (other.citatel * self.jmenovatel)

    def __ge__(self, other):
        return (self.citatel * other.jmenovatel) >= (other.citatel * self.jmenovatel)
z1 = Zlomek(1, 2)
z2 = Zlomek(2, -3)
z3 = Zlomek(2, 4)

print(z1)
print(z2)
print(z1 + z2)
print(z1 * z2)
print(z3.to_float())
print(z1 == z3)
print(z1 > z2)
````
