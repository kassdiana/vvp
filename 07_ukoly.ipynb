## 1.1 Základy
1. Vykreslete graf funkce $f(x) = \sin(x)$ na intervalu $x \in \langle -2\pi, 2\pi \rangle$.
- Graf vykreslete červenou čarou s tloušťkou 3.
- Velikost obrázku nastavte na 10x5 palců.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 500)
f_x = np.sin(x)
g_x = 100 ** np.sin(x)

fig, ax1 = plt.subplots(figsize=(10, 5))

color1 = 'tab:blue'
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color=color1)
ax1.plot(x, f_x, color=color1, label='sin(x)')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('100^sin(x)', color=color2)
ax2.plot(x, g_x, color=color2, label='100^sin(x)')
ax2.tick_params(axis='y', labelcolor=color2)

fig.tight_layout()
plt.show()
```

3. Pomocí `pcolormesh` vykreslete funkci $f(x, y) = \sin(x)\sin(y)$ na obdélníku $x \in \langle 0, 2\pi \rangle$, $y \in \langle 0, 2\pi \rangle$. Přidejte 5 obrysových čar pomocí `contour`.

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.figure(figsize=(8, 6))
mesh = plt.pcolormesh(X, Y, Z, cmap='viridis', shading='auto')
plt.colorbar(mesh, label='f(x, y)')
plt.contour(X, Y, Z, levels=5, colors='black', alpha=0.7)

plt.show()
```

## 1.6 3D grafy
1. Vykreslete 3D wireframe graf funkce
$$z = \frac{\sin(\sqrt{x^2 + y^2})}{\sqrt{x^2 + y^2}}.$$

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) / (R + 1e-10)

#wireframe
ax.plot_wireframe(X, Y, Z, rstride=3, cstride=3, color='teal', linewidth=0.7)
plt.title('3D Wireframe Graf')
plt.show()
```

2. Vytvořte 3D povrchový (surface) graf s funkcí $$z = \cos(\sqrt{x^2 + y^2}).$$

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.cos(R)

#surface
surf = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title('3D Surface Graf')
plt.show()
```
